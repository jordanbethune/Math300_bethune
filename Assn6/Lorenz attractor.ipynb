2.1 Write a function lorenz

```python
def lorenz(x,y,z,dt):
    
    a = 10
    b = 28
    c = 8/3
    
    dx = dt * (a * (y-x))
    dy = dt * (x * (b-z)-y) 
    dz = dt * ((x * y)-(c * z) )
    
    
    return (x + dx*dt,y + dy*dt,z + dz*dt) 
```

```python
lorenz(1,1,1,10,28,8/3,0.1)
```

2.2 Evolve lorenz for 1000 steps

```python
def lorenz(x,y,z):
    a = 10
    b = 28
    c = 8/3
    dt = .01
    
    results = []
    
    for i in range(1000):
        dx = (a * (y - x))
        xnew = (x + (dx * dt))
        
        dy = (x * (b - z) - y)
        ynew = (y + (dy * dt))
        
        dz = ((x * y) - (c * z))
        znew = (z + (dz * dt))
        
        results.append((xnew,ynew,znew))
        (x,y,z) = (xnew,ynew,znew)
        
    return results
```

```python
lorenz(1,1,1)
```

2.3 Plot projection of the iterated sequence

```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz(x,y,z):
    a = 10
    b = 28 
    c = 8/3
    dt = .01
    
    results = []
    
    for i in range(1000):
        dx = (a * (y - x))
        xnew = (x + (dx * dt))
        
        dy = (x * (b - z) - y)
        ynew = (y + (dy * dt))
        
        dz = ((x * y) - (c * z))
        znew = (z + (dz * dt))
        
        results.append((xnew,ynew,znew))
        (x,y,z) = (xnew,ynew,znew)
        
    return results
```

```python
res=lorenz(1,1,1)
```

```python
x = np.array(res)
```

```python
plt.plot(x[:,0],x[:,1])
plt.show()
```
